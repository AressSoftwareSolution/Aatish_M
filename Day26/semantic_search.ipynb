```python
# -----------------------------
# STEP 1: Imports
# -----------------------------
import chromadb
from chromadb.utils import embedding_functions
```

```python
# -----------------------------
# STEP 2: Sample Documents
# -----------------------------
documents = [
    "Exercise improves brain function and memory.",
    "Eating healthy food helps in weight management.",
    "Meditation reduces stress and improves focus.",
    "Strength training builds muscle mass.",
    "Reading books enhances knowledge and vocabulary.",
    "Drinking water is essential for good health."
]
```

```python
# We are storing all document IDs in a list called ids.
# Each ID corresponds to one document.
ids = [str(i) for i in range(len(documents))]
```

```python
# -----------------------------
# STEP 3: Create Chroma Client
# -----------------------------
client = chromadb.Client()  # creates a connection object to the Chroma database.

# Delete old collection if exists (for notebook reruns)
try:
    client.delete_collection("semantic_search_collection")
except:
    pass  # means Do nothing
```

```python
# -----------------------------
# STEP 4: Create Embedding Function
# -----------------------------
embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2" #It is a pre-trained embedding model
)

```

```python
# -----------------------------
# STEP 5: Create Collection
# -----------------------------
# Creates a new collection inside the vector database.
collection = client.create_collection(
    name="semantic_search_collection",  # This gives your collection a name.
    # It tells chroma "Whenever I add text or query text, use this embedding function to convert text into vectors".
    embedding_function=embedding_function
)

print("Collection created successfully.")

```

```python
# -----------------------------
# STEP 6: Add Documents
# -----------------------------
collection.add(
    documents=documents,
    ids=ids
)

print("Documents stored successfully.")
```

```python
# -----------------------------
# STEP 7: Query the System
# -----------------------------

# This method:

#   1.Converts the query into an embedding

#   2.Compares it with all stored embeddings

#   3.Calculates similarity

#   4.Returns the closest matches
query = "How to build muscle?"

results = collection.query(
    query_texts=[query],
    n_results=3
)

print("\nTop Results:\n")
for i, doc in enumerate(results["documents"][0]):
    print(f"{i+1}. {doc}")
```
